#Buscando Hotéis adaptados para PCD
###Este código implementa um Chat para busca de hotéis que possuam instalações adaptadas para pessoas com deficiencia (PCD).
O usuário entra com o nome da cidade e recebe uma lista dos hoteis com as seguimtes informações:

**- Nome do Hotel**

**- Bairro**

**- Site**

Ele poderá realizar varias pesquisas e ao terminar deverá digitar "fim".

```python
#Instalando o SDK do Google
!pip install -q -U google-generativeai
```

```python
#Configurações iniciais
import google.generativeai as genai

GOOGLE_API_KEY="INSIRA_SUA_API_KEY"
genai.configure(api_key=GOOGLE_API_KEY)
```

```python
#Configurando os parâmetros de avaliação dos tokens
generation_config = {
  "candidate_count": 1,
  "temperature": 1,
}
```

```python
#Configurando os parâmetros de segurança quanto ao conteudo da resposta
safety_settings={
    'HATE': 'BLOCK_NONE',
    'HARASSMENT': 'BLOCK_NONE',
    'SEXUAL' : 'BLOCK_NONE',
    'DANGEROUS' : 'BLOCK_NONE'
    }
```

```python
#Definindo o modelo utilizado
model = genai.GenerativeModel(model_name='gemini-1.0-pro',
                                  generation_config=generation_config,
                                  safety_settings=safety_settings,)
```

```python
chat = model.start_chat(history=[])

cidade = input('Qual a cidade que você vai viajar? Há, para terminar a pesquisa digite "fim". ')
prompt = 'Listar hoteis na cidade'+ cidade + 'com acesso a pessos com deficiencia.'

formatoResponse = '''Listar hoteis na cidade do Rio de Janeiro com acesso a pessos com deficiencia.
Nome do Hotel:  Copacabana Palace
Bairro: Copacabana
Site: https://hotelcopacabanapalace.com/
'''
response = chat.send_message(formatoResponse)

while prompt != "fim":
  print("")
  print("-------------------- Início lista ---------------------")
  print("")
  response = chat.send_message(prompt)
  print("Resposta:", response.text, '\n\n')
  prompt = input('Quer consultar outra cidade? Não se esqueça, para terminar digite "fim"')
```
